```python
import cv2
import matplotlib.pyplot as plt # Para exibir as imagens no Colab
```

```python
imagem = cv2.imread('pessoas-predios.jpg')
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
limiar = 127  # Escolha um valor de limiar (teste diferentes valores)
valor_maximo = 255

_, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, valor_maximo, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(15,5)) # Para ajustar o tamanho da exibição
plt.subplot(1,3,1) # 1 linha, 3 colunas, primeiro subplot
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')

plt.subplot(1,3,2)
plt.imshow(imagem_cinza, cmap='gray') # cmap='gray' para exibir em tons de cinza
plt.title('Imagem em Cinza')

plt.subplot(1,3,3)
plt.imshow(imagem_binarizada, cmap='gray')
plt.title('Imagem Binarizada')

plt.show()
```

```python
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_binarizada.jpg', imagem_binarizada)
```
